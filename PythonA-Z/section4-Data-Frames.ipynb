### ---Introduction to Seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.rcParams['figure.figsize'] = 8,4
```

```python
data.head()
```

```python
#Distribution of Internet  users
viz = sns.distplot(data['InternetUsers'])
```

```python
viz1 = sns.distplot(data['InternetUsers'], bins=10)
```

```python
#Boxplots:
viz2 = sns.boxplot(data=data, x="IncomeGroup", y="BirthRate")
```

```python
viz3 =  sns.lmplot(data=data, x="InternetUsers", y="BirthRate", fit_reg=False)
```

```python
viz4 =  sns.lmplot(data=data, x="InternetUsers", y="BirthRate", fit_reg=False, hue="IncomeGroup", height=7)
```

```python
## Keyword Arguments in Python
```

```python
#Marker size
viz4 =  sns.lmplot(data=data, x="InternetUsers", y="BirthRate", fit_reg=False, hue="IncomeGroup", height=7, 
                  scatter_kws={'s':100})
```
